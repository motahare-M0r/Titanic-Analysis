<h1>Library</h1>

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

<h1>Input</h1>

```python
titanic = pd.read_csv("titanic.csv")
print(titanic)
```

<h1>Data Understanding</h1>

```python
titanic.info()
```

```python
titanic.isna().sum()
```

```python
titanic[["sex" , "survived"]].value_counts()
```

<h1>Visualization</h1>

```python
#Average of the living and the dead

sns.barplot(data=titanic, x="sex", y="survived", palette=["blue" , "pink"])
plt.title("Average of the living and the dead")
plt.show()
```

```python
# 1 : Living and dead people of different genders

peapole = titanic[["sex", "survived"]].value_counts().reset_index(name="count")

plt.bar(range(len(peapole)), peapole["count"], color="red")
plt.xticks(range(len(peapole)), peapole["sex"] + " - " + peapole["survived"].astype(str))
plt.xlabel("Groups")
plt.ylabel("Count")
plt.title("Living and dead people of different genders")
plt.show()
```

```python
print(peapole)
```

```python
# 2 : Living and dead people of different genders

sns.countplot(data=titanic, x="survived", hue="sex")

plt.xlabel("0 = dead  ,  1 = living")
plt.ylabel("Count")
plt.title("Living and dead people of different genders")
plt.legend(["female", "male"])
plt.show()
```
